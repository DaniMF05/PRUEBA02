### Interpole los siguientes conjuntos de datos con la función correspondiente. 

La ecuación de la línea es:

$ y(x) = a_1 x + a_0 $

Al realizar el proceso de mínimos cuadrados queda el siguiente sistema de ecuaciones:


$( \sum_i (y_i - a_1 x_i - a_0) , \sum_i (y_i - a_1 x_i - a_0)x_i ) = 0$

```python
# Derivadas parciales para regresión cuadrática
# ####################################################################

def der_parcial_0(xs: list, ys: list) -> tuple[float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro a_0 
    para una función cuadrática y = a_2*x^2 + a_1*x + a_0.
    La ecuación es de la forma: c_2 * a_2 + c_1 * a_1 + c_0 * a_0 = c_ind

    ## Parameters
    ``xs``: lista de valores de x.
    ``ys``: lista de valores de y.

    ## Return
    ``c_2``: coeficiente del parámetro a_2.
    ``c_1``: coeficiente del parámetro a_1.
    ``c_0``: coeficiente del parámetro a_0.
    ``c_ind``: coeficiente del término independiente.
    """
    n = len(xs)
    
    # Coeficientes de la ecuación: sum(a_0) + sum(a_1*x_i) + sum(a_2*x_i^2) = sum(y_i)
    c_0 = n  # coeficiente de a_0
    c_1 = sum(xs)  # coeficiente de a_1
    c_2 = sum(xi * xi for xi in xs)  # coeficiente de a_2
    c_ind = sum(ys)  # término independiente
    
    return (c_2, c_1, c_0, c_ind)


def der_parcial_1(xs: list, ys: list) -> tuple[float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro a_1 
    para una función cuadrática y = a_2*x^2 + a_1*x + a_0.
    La ecuación es de la forma: c_2 * a_2 + c_1 * a_1 + c_0 * a_0 = c_ind

    ## Parameters
    ``xs``: lista de valores de x.
    ``ys``: lista de valores de y.

    ## Return
    ``c_2``: coeficiente del parámetro a_2.
    ``c_1``: coeficiente del parámetro a_1.
    ``c_0``: coeficiente del parámetro a_0.
    ``c_ind``: coeficiente del término independiente.
    """
    
    # Coeficientes de la ecuación: sum(a_0*x_i) + sum(a_1*x_i^2) + sum(a_2*x_i^3) = sum(y_i*x_i)
    c_0 = sum(xs)  # coeficiente de a_0
    c_1 = sum(xi * xi for xi in xs)  # coeficiente de a_1
    c_2 = sum(xi * xi * xi for xi in xs)  # coeficiente de a_2
    c_ind = sum(yi * xi for xi, yi in zip(xs, ys))  # término independiente
    
    return (c_2, c_1, c_0, c_ind)


def der_parcial_2(xs: list, ys: list) -> tuple[float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro a_2 
    para una función cuadrática y = a_2*x^2 + a_1*x + a_0.
    La ecuación es de la forma: c_2 * a_2 + c_1 * a_1 + c_0 * a_0 = c_ind

    ## Parameters
    ``xs``: lista de valores de x.
    ``ys``: lista de valores de y.

    ## Return
    ``c_2``: coeficiente del parámetro a_2.
    ``c_1``: coeficiente del parámetro a_1.
    ``c_0``: coeficiente del parámetro a_0.
    ``c_ind``: coeficiente del término independiente.
    """
    
    # Coeficientes de la ecuación: sum(a_0*x_i^2) + sum(a_1*x_i^3) + sum(a_2*x_i^4) = sum(y_i*x_i^2)
    c_0 = sum(xi * xi for xi in xs)  # coeficiente de a_0
    c_1 = sum(xi * xi * xi for xi in xs)  # coeficiente de a_1
    c_2 = sum(xi * xi * xi * xi for xi in xs)  # coeficiente de a_2
    c_ind = sum(yi * xi * xi for xi, yi in zip(xs, ys))  # término independiente
    
    return (c_2, c_1, c_0, c_ind)
```
